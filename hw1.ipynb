idea:
1. 时间复杂度递归式及渐近表示
设 n = r - p + 1，表示当前需要处理的数组区间长度。
当 n = 2 时，算法只比较 A[p] 和 A[r]，如果顺序错误则交换。
因此该情况只需要常数时间：
    T(2) = O(1)
当 n >= 3 时：
    m = ceil(2n / 3)
算法会进行三次递归调用：
    DoSomething(A, p, p + m - 1)
    DoSomething(A, r - m + 1, r)
    DoSomething(A, p, p + m - 1)
每一次递归调用的子问题规模约为 2n / 3，
并且除了递归调用之外，只进行了赋值和判断等常数时间操作。
因此递归式为：
    T(n) = 3T(ceil(2n / 3)) + O(1)
忽略向上取整对渐近复杂度的影响，可以写成：
    T(n) = 3T(2n / 3) + O(1)
下面使用递归树进行分析。
第 0 层有 1 个子问题；
第 1 层有 3 个子问题；
第 2 层有 3^2 个子问题；
第 i 层有 3^i 个子问题。
第 i 层每个子问题的规模为：
    (2 / 3)^i * n
当子问题规模变为常数时递归结束：
    (2 / 3)^k * n = 1
所以：
    k = log_{3/2} n
总时间复杂度为：
    T(n) = 1 + 3 + 3^2 + ... + 3^k
         = O(3^k)
将 k = log_{3/2} n 代入：
    T(n) = O(3^(log_{3/2} n))
         = O(n^(log_{3/2} 3))
因为 log_{3/2} 3 约等于 2.71，所以：
    T(n) = O(n^2.71)
因此，该算法的渐近时间复杂度为：
    O(n^(log_{3/2} 3))

2. 算法功能及是否高效
该算法的功能是对数组 A[p:r] 进行升序排序。
这个算法实际上是 Stooge Sort，也叫臭皮匠排序。
它的基本思想是：
    1. 如果当前区间只有两个元素，则比较并交换；
    2. 如果当前区间至少有三个元素，则先排序前 2/3 部分；
    3. 然后排序后 2/3 部分；
    4. 最后再次排序前 2/3 部分。
通过不断递归地处理重叠的前 2/3 和后 2/3 区间，
算法最终可以将整个数组排成升序。
但是，该算法不是最高效的解决方案。
它的时间复杂度为：
    O(n^(log_{3/2} 3)) ≈ O(n^2.71)
而常见的高效排序算法，例如归并排序、堆排序、快速排序，
可以达到：
    O(n log n)
因此，该算法虽然能够完成排序任务，但效率较低，
不是最高效的排序方案。

````python
def merge_sort(A):
    """
    归并排序：将数组 A 原地升序排序
    """
    temp = [0] * len(A)

    def sort(left, right):
        # 排序区间 A[left:right]
        if right - left <= 1:
            return

        mid = (left + right) // 2

        sort(left, mid)
        sort(mid, right)

        i = left
        j = mid
        k = left

        while i < mid and j < right:
            if A[i] <= A[j]:
                temp[k] = A[i]
                i += 1
            else:
                temp[k] = A[j]
                j += 1
            k += 1

        while i < mid:
            temp[k] = A[i]
            i += 1
            k += 1

        while j < right:
            temp[k] = A[j]
            j += 1
            k += 1

        for t in range(left, right):
            A[t] = temp[t]

    sort(0, len(A))
    return A


# example
A = [5, 3, 8, 1, 2, 7]
merge_sort(A)
print(A)

# your algorithm time complexity is:
# Θ(n log n)
````

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:1. 算法方案分析 （动态规划 / Kadane's Algorithm） 由于总收益等于连续子数组的和，我们需要在数组中找到一个连续的区间，使得该区间的元素之和最大。
我们可以使用动态规划（具体为 Kadane 算法）来解决这个问题：状态定义：维护一个变量 ，表示以当前第k天结尾的连续子数组的最大和。 同时维护一个 记录全局找到的最大和。
状态转移：对于第k天的收益 ，我们面临两个选择：profits[k]将其加入之前的连续区间中：current_sum + profits[k]放弃之前的区间，从第k天重新开始计算： 因此，状态转移方程为：profits[k] current_sum = max(profits[k], current_sum + profits[k])记录日期：为了记录买入和卖出日期，我们需要在 重新开始计算时（即），记录下新的起始点 。 当打破全局记录更新 时，把全局的起始点 更新为 ，结束点 更新为当前索引 current_sumprofits[k] > current_sum + profits[k]temp_startcurrent_summax_sumstarttemp_startend$k$。

````python
def max_profit_subarray(profits):
    # 处理边界情况：空数组
    if not profits:
        return 0, -1, -1

    # 初始化变量
    max_sum = profits[0]       # 全局最大收益
    current_sum = profits[0]   # 当前连续区间的收益
    
    start_idx = 0              # 全局最大收益的买入索引
    end_idx = 0                # 全局最大收益的卖出索引
    temp_start = 0             # 当前连续区间的潜在买入索引

    # 遍历数组（从第二个元素开始）
    for i in range(1, len(profits)):
        # 如果当前收益加上前面的和反而不如当前收益单干（即前面的和为负数）
        # 则抛弃前面的区间，从当前节点重新开始计算买入
        if current_sum + profits[i] < profits[i]:
            current_sum = profits[i]
            temp_start = i
        else:
            current_sum += profits[i]

        # 如果当前区间的和大于全局最大和，则更新记录和具体的买卖日期
        if current_sum > max_sum:
            max_sum = current_sum
            start_idx = temp_start
            end_idx = i

    # 题目示例中天数是基于 1 开始的（1-indexed），而数组索引是基于 0 的
    # 所以返回时需要将索引 + 1 转化为具体天数
    buy_day = start_idx + 1
    sell_day = end_idx + 1

    return max_sum, buy_day, sell_day

# add your code here
# 测试用例
profits_example = [3, 2, 1, -7, 5, 2, -1, 3, -1]
max_profit, buy_day, sell_day = max_profit_subarray(profits_example)

print(f"profits={profits_example}, 第{buy_day}天买入, 第{sell_day}天卖出, 收益最大: {max_profit}")

# your algorithm time complexity is: O(n)
# your algorithm space complexity is: O(1)
````

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:1. 寻找一个局部最小（时间复杂度 $O(\log n)$）设计思路 （二分查找）：
题目给出了两个非常关键的边界条件：$A[0] > A[1]$（开头呈下降趋势）和 $A[n-2] < A[n-1]$（结尾呈上升趋势）。 因为数组中没有重复元素，一个先下降后上升的连续序列中，必定存在至少一个“波谷”（也就是局部最小值）。
我们可以利用二分查找：取数组中间的元素 $A[mid]$。命中目标：如果它比左右两边的元素都小（$A[mid] < A[mid-1]$ 且 $A[mid] < A[mid+1]$），那么它就是我们要找的局部最小，直接返回。向右侧寻找：如果 $A[mid] > A[mid+1]$，说明当前在 这个位置趋势是向下的。 因为数组末尾必定是向上的，所以在 到末尾之间一定藏着一个波谷。 我们将搜索区间缩小到右半部分（）。midmidleft = mid + 1向左侧寻找：如果上述条件都不满足（此时必然有 $A[mid] > A[mid-1]$），说明当前在 位置趋势是向上的。 结合数组开头是向下的，所以在开头到 之间一定有一个波谷。 我们将搜索区间缩小到左半部分（）。midmidright = mid - 12. 寻找所有局部最小值设计思路 （线性遍历）：
要找到所有的局部最小值，我们无法依赖 $O(\log n)$ 的算法，因为在最坏情况下（比如数组是锯齿状的），局部最小值的个数可能达到 $\frac{n}{2}$ 个。 只要需要输出的数量级是 $O(n)$，算法的时间复杂度就必定至少是 $O(n)$。
因此，最简单且最优解就是进行一次线性扫描。 由于题目的边界条件保证了第一个和最后一个元素不可能是局部最小值，我们只需要遍历索引 $1$ 到 $n-2$ 的元素，判断它是否同时小于左右邻居即可。

````python
# add your code here
def find_one_local_minimum(A):
    """
    问题 1: 在 O(log n) 时间复杂度下找到一个局部最小值
    """
    n = len(A)
    # 题目保证了 A[0] > A[1] 且 A[n-2] < A[n-1]，所以边界不可能是局部最小
    # 我们的搜索范围可以在 1 到 n-2 之间
    left, right = 1, n - 2
    
    while left <= right:
        mid = (left + right) // 2
        
        # 判断是否为局部最小
        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]
        
        # 如果当前呈下降趋势，波谷必定在右侧
        elif A[mid] > A[mid + 1]:
            left = mid + 1
            
        # 如果当前呈上升趋势，波谷必定在左侧
        else:
            right = mid - 1
            
    return -1 # 理论上根据题目条件不会走到这里


def find_all_local_minima(A):
    """
    问题 2: 找到所有的局部最小值
    """
    n = len(A)
    minima_list = []
    
    # 线性遍历，避开两端元素
    for i in range(1, n - 1):
        if A[i] < A[i - 1] and A[i] < A[i + 1]:
            minima_list.append(A[i])
            
    return minima_list

# 测试用例
A_example = [9, 3, 7, 2, 1, 4, 5]

# 1. 测试找到一个局部最小
one_min = find_one_local_minimum(A_example)
print(f"找到的一个局部最小值为: {one_min}  (时间复杂度: O(log n))")

# 2. 测试找到所有局部最小
all_min = find_all_local_minima(A_example)
print(f"所有的局部最小值列表为: {all_min}  (时间复杂度: O(n))")
# your algorithm time complexity is:
````

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:1. 基于排序的方法 （全局排序）设计思路：这是最直观的方法。 直接将包含 $n$ 个元素的数组进行升序排序，然后直接截取并输出前 $k$ 个元素即可。时间复杂度计算：通常采用高效的比较排序算法（如归并排序、快速排序或 Python 内置的 Timsort），对 $n$ 个元素排序的时间复杂度为 $O(n \log n)$。 截取前 $k$ 个元素的时间复杂度为 $O(k)$。 因此，总时间复杂度为 $O(n \log n)$。2. 基于最小堆的方法 （Min-Heap）设计思路：利用堆这种数据结构。 题目要求“往空堆中不断插入元素”，即遍历数组，将 $n$ 个元素逐一插入到最小堆中。 建堆完成后，执行 $k$ 次 （弹出堆顶元素）操作，依次得到的也就是最小的 Extract-min$k$ 个数。时间复杂度计算：建堆：向堆中插入第 $i$ 个元素时，最多需要调整 $\log i$ 次。 连续插入 $n$ 个元素，建堆的时间复杂度为 $\sum_{i=1}^{n} \log i = O(n \log n)$。 （注：如果使用线性建堆法可以达到 $O(n)$，但题目限定了逐个插入的方法）。提取：每次执行 并重新调整堆的时间为 Extract-min$O(\log n)$，$k$ 次提取的时间复杂度为 $O(k \log n)$。总时间复杂度：$O(n \log n + k \log n) = O(n \log n)$。3. 处理海量数据的方法 （Max-Heap）设计思路：当数据量 $n$ 超出内存，但内存可以放下 $k$ 个数据时，我们不能把所有数据读入内存建最小堆。 此时的核心思路是维护一个容量为 $k$ 的最大堆 （Max-Heap）。读取数据流的前 $k$ 个元素，建立一个容量为 $k$ 的最大堆。 此时堆顶是这 $k$ 个元素中的最大值。继续读取后续的每一个元素 $x$。 将其与最大堆的堆顶进行比较：如果 $x \ge$ 堆顶元素，说明 $x$ 绝不可能是全局最小的 $k$ 个数之一，直接丢弃。如果 $x <$ 堆顶元素，说明堆顶元素被淘汰了。 我们将堆顶弹出，并将 $x$ 插入堆中。数据读取完毕后，最大堆里留下的就是全局最小的 $k$ 个数。 最后将这 $k$ 个数进行逆序（或排序）输出即可。时间复杂度计算：维护一个容量为 $k$ 的堆，每次插入/弹出的时间复杂度为 $O(\log k)$。 最坏情况下（数据是降序的），对于 $n$ 个元素都要操作堆，总时间复杂度为 $O(n \log k)$。 这不仅完美解决了内存限制问题，且在 $k \ll n$ 时，效率远高于 $O(n \log n)$。 空间复杂度为 $O(k)$。

````python
# add your code here
import heapq

# ==================== 问题 1：全局排序法 ====================
def get_least_k_by_sorting(arr, k):
    if not arr or k <= 0:
        return []
    # 使用 Python 内置排序 (Timsort)，时间复杂度 O(n log n)
    sorted_arr = sorted(arr) 
    return sorted_arr[:k]

# ==================== 问题 2：最小堆法 ====================
def get_least_k_by_min_heap(arr, k):
    if not arr or k <= 0 or k > len(arr):
        return []
    
    min_heap = []
    # 按照题意，往空堆中不断插入元素建堆。时间复杂度 O(n log n)
    for num in arr:
        heapq.heappush(min_heap, num)
        
    result = []
    # 调用 k 次 Extract-min 按顺序输出最小的 k 个数。时间复杂度 O(k log n)
    for _ in range(k):
        result.append(heapq.heappop(min_heap))
        
    return result

# ==================== 问题 3：海量数据/最大堆法 ====================
# 假设 arr 是一个无法一次性读入内存的数据流（这里用列表模拟迭代器）
def get_least_k_by_max_heap(data_stream, k):
    if k <= 0:
        return []
        
    max_heap = []
    
    # 模拟从海量数据流中逐个读取数据
    for i, num in enumerate(data_stream):
        if i < k:
            # 前 k 个元素，取反后压入堆，模拟最大堆
            heapq.heappush(max_heap, -num)
        else:
            # 从第 k+1 个元素开始，与堆顶（当前 k 个数中的最大值）比较
            # max_heap[0] 是负数，-max_heap[0] 就是真实的堆顶最大值
            if num < -max_heap[0]:
                # 如果当前元素比堆顶小，则弹出堆顶，压入新元素
                heapq.heapreplace(max_heap, -num)
                
    # 此时 max_heap 中保存了最小的 k 个数的负值
    # 题目要求按顺序输出，所以取出后取反，并进行升序排序
    result = [-x for x in max_heap]
    result.sort() # 对 k 个元素排序的时间复杂度为 O(k log k)
    
    return result

# ==================== 测试代码 ====================
if __name__ == "__main__":
    arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
    k = 3
    
    print(f"原始数组: {arr}, k={k}\n")
    
    res1 = get_least_k_by_sorting(arr, k)
    print(f"1. 排序法结果: {res1}  (Time: O(n log n))")
    
    res2 = get_least_k_by_min_heap(arr, k)
    print(f"2. 最小堆结果: {res2}  (Time: O(n log n))")
    
    res3 = get_least_k_by_max_heap(arr, k)
    print(f"3. 最大堆结果: {res3}  (Time: O(n log k))")
# your algorithm time complexity is:
````

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:1. 算法设计思路 （栈的应用） 星际碰撞只发生在一种特定情况下：一个向右移动的星球（正数）位于一个向左移动的星球（负数）的左侧。  因此，我们可以使用一个栈 （Stack） 来模拟星球的运行轨迹：遍历给定的星球数组。对于当前遍历到的星球 ，我们判断是否会发生碰撞：p不会碰撞的情况：如果栈为空，或者当前星球 是向右移动的（），或者栈顶的星球也是向左移动的（）。 此时，直接将 入栈。pp > 0stack[-1] < 0p会发生碰撞的情况：如果栈顶星球向右移动（）且当前星球 向左移动（），则必然相撞。stack[-1] > 0pp < 0当发生相撞时，比较质量大小（绝对值）：如果 ：栈顶星球质量小，被毁灭（出栈），当前星球 继续与新的栈顶判断是否会相撞。stack[-1] < abs(p)p如果 ：当前星球 质量小，被毁灭，停止比较， 不入栈。stack[-1] > abs(p)pp特殊规则 ：质量相同，题意规定“自左向右的保持不变，自右向左的消失”。 这意味着栈顶的正数星球保留，当前负数星球 消失。 停止比较， 不入栈。stack[-1] == abs(p)pp2. 时间复杂度与空间复杂度时间复杂度：$O(n)$。 虽然代码中包含了一个 循环，但每个星球最多只会入栈一次，也最多只会被弹出（出栈）一次。 因此，平摊到每个元素上的操作是常数级别的。while空间复杂度：$O(n)$。 在最坏的情况下（例如所有的星球都是向右移动的，或者所有向左的都在最左边），没有星球会发生碰撞，全部存入栈中。

````python
# add your code here
def asteroid_collision(A):
    stack = []
    
    for p in A:
        # alive 标志当前遍历的星球 p 是否存活
        alive = True
        
        # 当发生碰撞的条件满足时：栈不为空 且 栈顶向右(>0) 且 当前星球向左(<0)
        while alive and p < 0 and stack and stack[-1] > 0:
            # 情况 1: 栈顶星球质量小于当前星球
            if stack[-1] < abs(p):
                stack.pop() # 栈顶星球被毁灭
                continue    # 当前星球 p 继续与下一个栈顶比较
                
            # 情况 2: 质量相同 (题目特殊规则)
            elif stack[-1] == abs(p):
                alive = False # 自右向左的星球(p)消失，栈顶保持不变
                
            # 情况 3: 栈顶星球质量大于当前星球
            else:
                alive = False # 当前星球 p 被毁灭
                
        # 如果当前星球历经碰撞后仍然存活，或者压根没发生碰撞，则入栈
        if alive:
            stack.append(p)
            
    return stack

# add your code here
# 测试用例 1
A1 = [-3, -6, 2, 8, 5, -8, 9, -2, 1]
print(f"输入: A={A1}")
print(f"输出: {asteroid_collision(A1)}")

print("-" * 20)

# 测试用例 2
A2 = [23, -8, 9, -3, -7, 9, -23, 22]
print(f"输入: A={A2}")
print(f"输出: {asteroid_collision(A2)}")

# your algorithm time complexity is: O(n)

````

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:1. 算法设计思路 （动态规划 Dynamic Programming） 解决 0-1 背包问题的核心是使用动态规划。状态定义：我们定义一个一维数组 ，其中 表示当推车容量为 时，所能装下的最大大理石价值。dpdp[j]j初始化：初始时，没有任何物品，所以 数组的所有元素都初始化为 0。dp状态转移方程：
遍历每一个大理石（获取其重量 和价值 ）。 对于当前的大理石，我们遍历推车的容量 。
如果不把这块大理石装入车里，那么最大价值依然是之前的 ;
如果把这块大理石装入车里（前提是当前容量 足够大，即 ），那么最大价值就变成了 。
我们需要在“装”与“不装”之间选择一个最大值，因此状态转移方程为：wvjdp[j]jj >= wdp[j - w] + v$$dp[j] = \max(dp[j], dp[j - w] + v)$$⚠️ 关键点（逆序遍历）：在更新一维 数组时，内层循环（容量 ）必须从最大容量向小容量逆序遍历（即从 递减到 ）。 这是为了保证在计算 时，它所依赖的 是上一轮的状态，从而确保每块大理石最多只被放入一次。dpjcapacitywdp[j]dp[j - w]2. 时间与空间复杂度计算时间复杂度：$O(N \cdot W)$。 其中 $N$ 是大理石的数量（此题为 4），$W$ 是推车的最大承重（此题为 9）。 我们需要两层嵌套循环，外层遍历 $N$ 个物品，内层遍历最大为 $W$ 的容量。空间复杂度：$O(W)$。 由于我们通过状态压缩将原本二维的动态规划表优化成了一维数组，只需要一个长度为 $W + 1$ 的数组即可，因此空间复杂度为 $O(W)$。

````python
# add your code here
def knapsack_01(items, capacity):
    """
    0-1背包问题求解
    :param items: list of tuples, 每个元组为 (重量, 价值)
    :param capacity: int, 背包(小推车)的最大承重
    :return: int, 最大总价值
    """
    # dp[j] 表示容量为 j 时所能装下的最大价值
    # 初始化为 0，长度为 capacity + 1 (包含容量为0的情况)
    dp = [0] * (capacity + 1)
    
    # 遍历每一块大理石
    for weight, value in items:
        # 必须逆序遍历容量，保证每个物品只能被选择一次
        for j in range(capacity, weight - 1, -1):
            # 状态转移：选择“不装”当前物品和“装入”当前物品中价值更大的一个
            dp[j] = max(dp[j], dp[j - weight] + value)
            
    # dp 数组的最后一个元素就是最大承重下的最大价值
    return dp[capacity]

# add your code here
# 测试用例：题目给定的数据
marbles = [(5, 10), (4, 40), (6, 30), (3, 50)]
cart_capacity = 9

max_value = knapsack_01(marbles, cart_capacity)

print(f"大理石清单 (重量, 价值): {marbles}")
print(f"小推车最大承重: {cart_capacity}")
print(f"能装载的最大价值为: {max_value}")

# your algorithm time complexity is: O(N*W)

````

我们只需要在中序遍历的过程中，记录下前一个访问的节点（），并将其与当前访问的节点（）进行比较：如果 ，说明我们找到了一个逆序对。prevcurrprev.val > curr.val根据被交换的两个节点在序列中的位置，会出现两种情况：情况 A（相邻交换）：比如序列 被换成了 。 此时只会产生一个逆序对 。 被交换的节点就是这个逆序对的两个节点。[1, 2, 3, 4, 5][1, 3, 2, 4, 5](3, 2)情况 B（不相邻交换）：比如序列 被换成了 。 此时会产生两个逆序对 和 。 被交换的节点是第一个逆序对的第一个节点（5）和第二个逆序对的第二个节点（2）。[1, 2, 3, 4, 5][1, 5, 3, 4, 2](5, 3)(4, 2)统一处理逻辑：维护三个指针：（第一个错误节点），（第二个错误节点），（前驱节点）。firstsecondprev进行中序遍历。 当遇到 时：prev.val > curr.val如果 为空，说明是第一次遇到逆序对，将 赋给 ，将 赋给 。firstprevfirstcurrsecond如果 不为空，说明是第二次遇到逆序对，直接将 赋给 即可（覆盖掉之前的值）。firstcurrsecond遍历结束后，直接交换 和 节点的值，树的结构不变，从而恢复二叉搜索树。firstsecond2. 时间与空间复杂度计算时间复杂度：$O(n)$。 在最坏情况下，我们需要遍历二叉树中的所有节点一次，其中 $n$ 为二叉树的节点数。空间复杂度：$O(h)$。 主要为递归调用栈的开销，其中 $h$ 是树的高度。 最坏情况下（树退化为链表）空间复杂度为 $O(n)$，平均情况下为 $O(\log n)$。 （注：如果题目进阶要求 $O(1)$ 空间复杂度，可以使用 Morris 中序遍历算法，但通常标准的递归解法已满足要求且更具可读性。）

````python
# add your code here
# 定义二叉树节点
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def recoverTree(root: TreeNode) -> None:
    """
    Do not return anything, modify root in-place instead.
    """
    first = None
    second = None
    prev = None

    # 定义中序遍历的递归函数
    def inorder(node):
        nonlocal first, second, prev
        if not node:
            return
        
        # 1. 遍历左子树
        inorder(node.left)
        
        # 2. 处理当前节点
        if prev and prev.val > node.val:
            # 找到逆序对
            if not first:
                # 第一次遇到逆序对，记录第一个错误节点
                first = prev
            # 无论是不是第一次，都将当前节点记录为潜在的第二个错误节点
            second = node
            
        # 更新前驱节点为当前节点
        prev = node
        
        # 3. 遍历右子树
        inorder(node.right)

    # 从根节点开始中序遍历
    inorder(root)
    
    # 遍历完成后，交换找到的两个节点的值
    if first and second:
        first.val, second.val = second.val, first.val


# add your code here
# ==================== 测试用例 1 ====================
# 构造题图中的示例 1: root = [1,3,null,null,2]
#    1
#   /
#  3
#   \
#    2
root1 = TreeNode(1)
root1.left = TreeNode(3)
root1.left.right = TreeNode(2)

recoverTree(root1)
# 恢复后应为: root = [3,1,null,null,2]
# 打印根节点及其左子树验证
print(f"恢复后 示例1 根节点的值: {root1.val}")            # 应为 3
print(f"恢复后 示例1 左子节点的值: {root1.left.val}")      # 应为 1

# your algorithm time complexity is: O(n)
# your algorithm space complexity is: O(h)
````

idea:1. 算法设计思路 （双指针法） 要计算总共能接多少雨水，我们可以按列来计算：每一列（每一个柱子）上面能积攒的水量，取决于它左边最高的柱子和右边最高的柱子中，较矮的那一个。
即：。当前列水量 = max(0, min(左侧最高, 右侧最高) - 当前高度)如果使用常规方法，我们需要提前算好每个位置的左侧最高和右侧最高，空间复杂度是 $O(n)$。 使用双指针法可以将空间复杂度降为 $O(1)$：维护两个指针 和 ，分别指向数组的头和尾。leftright维护两个变量 和 ，分别记录左侧已经扫描过的最高柱子和右侧已经扫描过的最高柱子。left_maxright_max当 时：
这说明当前 柱子的右侧一定存在一个比 更高（或一样高）的天然屏障。 因此， 位置能接多少水，完全只取决于它左侧的最高点 left_max。height[left] < height[right]leftheight[left]left如果 ，说明当前柱子比左边最高的还高，装不了水，更新 。height[left] >= left_maxleft_max否则，当前柱子上方可以接水，累加 。left_max - height[left]将 指针右移。left反之，当 时：
说明 柱子的左侧一定存在一个足够高的屏障。 位置能接多少水，完全取决于它右侧的最高点 right_max。 处理逻辑同上，处理完后将 指针左移。height[left] >= height[right]rightrightright两指针相遇时，遍历结束。2. 时间与空间复杂度计算时间复杂度：$O(n)$。 两个指针 和 最多只会相遇一次，也就是遍历整个数组一次，每次操作的时间是常数级别。leftright空间复杂度：$O(1)$。 只使用了有限的几个额外变量（指针和最大值记录），没有开辟新的数组空间，达到了理论最优。

````python
# add your code here
def trap(height):
    """
    计算下雨之后能接多少雨水
    """
    if not height:
        return 0
        
    n = len(height)
    left, right = 0, n - 1
    left_max, right_max = 0, 0
    total_water = 0
    
    while left < right:
        # 哪边较低，哪边的盛水量就有了决定性的短板，可以直接计算
        if height[left] < height[right]:
            # 处理左侧指针
            if height[left] >= left_max:
                left_max = height[left] # 更新左侧最高屏障
            else:
                # 累加积水量
                total_water += left_max - height[left]
            left += 1
            
        else:
            # 处理右侧指针
            if height[right] >= right_max:
                right_max = height[right] # 更新右侧最高屏障
            else:
                # 累加积水量
                total_water += right_max - height[right]
            right -= 1
            
    return total_water

# add your code here
# ==================== 测试用例 ====================
# 示例 1
height1 = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
print(f"输入: height = {height1}")
print(f"输出: {trap(height1)}")  # 预期输出: 6

print("-" * 20)

# 示例 2
height2 = [4, 2, 0, 3, 2, 5]
print(f"输入: height = {height2}")
print(f"输出: {trap(height2)}")  # 预期输出: 9

# your algorithm time complexity is: O(n)
# your algorithm space complexity is: O(1)
````

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea:这是一道非常经典的二叉树路径求和问题，在 LeetCode 上对应第 129 题“求根节点到叶节点数字之和”。这道题的核心是利用深度优先搜索（DFS），自顶向下地遍历整棵树，并在向下遍历的过程中“顺藤摸瓜”地把数字拼接起来。以下是详细的算法分析和 Python 实现代码：idea:1. 算法设计思路 （深度优先搜索 DFS）状态传递：我们从根节点开始向下遍历。 在向下一层深入的过程中，维护一个变量 ，用来表示从根节点到当前节点路径上所形成的数字。current_num数字拼接规则：由于在十进制中，每向后追加一位数字，相当于把之前的数字乘以 10 再加上新数字。 因此，状态转移方程为：。current_num = current_num * 10 + node.val终止与结算条件：如果当前节点为空（走到了空分支），返回 0。如果当前节点是叶子节点（即左右子树都为空），说明一条完整的路径走完了，此时直接返回算好的 。current_num递归汇总：对于任何一个既有左子树又有右子树的中间节点，它负责把左边子树算出来的总和，加上右边子树算出来的总和，然后向上层返回。2. 时间与空间复杂度计算时间复杂度：$O(N)$。 其中 $N$ 是二叉树的节点总数。 深度优先搜索需要遍历二叉树中的每一个节点恰好一次。空间复杂度：$O(H)$。 其中 $H$ 是二叉树的高度。 空间复杂度主要取决于递归调用的隐式栈空间的最大深度。 在最坏情况下（树退化成一条链），高度 $H = N$，空间复杂度为 $O(N)$。 在平均情况下（平衡二叉树），高度 $H = \log N$，空间复杂度为 $O(\log N)$。

````python
# add your code here
# 定义二叉树节点类
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def sumNumbers(root: TreeNode) -> int:
    def dfs(node, current_num):
        # 如果节点为空，说明这条路径无效，和为 0
        if not node:
            return 0
            
        # 拼接当前节点的值
        current_num = current_num * 10 + node.val
        
        # 如果是叶子节点，返回当前路径组成的数字
        if not node.left and not node.right:
            return current_num
            
        # 否则，继续递归遍历左子树和右子树，并将它们的总和相加
        left_sum = dfs(node.left, current_num)
        right_sum = dfs(node.right, current_num)
        
        return left_sum + right_sum

    # 从根节点开始 DFS，初始数字为 0
    return dfs(root, 0)

# add your code here
# ==================== 测试用例 ====================

# 测试用例 1: root = [1, 2, 3]
root1 = TreeNode(1)
root1.left = TreeNode(2)
root1.right = TreeNode(3)
print(f"示例 1 输出: {sumNumbers(root1)}")  # 预期输出: 25

# 测试用例 2: root = [4, 9, 0, 5, 1]
root2 = TreeNode(4)
root2.left = TreeNode(9)
root2.right = TreeNode(0)
root2.left.left = TreeNode(5)
root2.left.right = TreeNode(1)
print(f"示例 2 输出: {sumNumbers(root2)}")  # 预期输出: 1026

# your algorithm time complexity is: O(N)
# your algorithm space complexity is: O(H)
````
