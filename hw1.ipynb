idea:

2. 该算法的功能是对数组 A[p:r] 进行升序排序。
这个算法实际上是 Stooge Sort，也叫臭皮匠排序。
它的基本思想是：
    （1）. 如果当前区间只有两个元素，则比较并交换；
    （2）. 如果当前区间至少有三个元素，则先排序前 2/3 部分；
    （3）. 然后排序后 2/3 部分；
    （4）. 最后再次排序前 2/3 部分。
通过不断递归地处理重叠的前 2/3 和后 2/3 区间，
算法最终可以将整个数组排成升序。
但是，该算法不是最高效的解决方案。
它的时间复杂度为：
    O(n^(log_{3/2} 3)) ≈ O(n^2.71)
而常见的高效排序算法，例如归并排序、堆排序、快速排序，
可以达到：
    O(nlogn)
因此，该算法虽然能够完成排序任务，但效率较低，
不是最高效的排序方案。

````python
def merge_sort(A):
    temp = [0] * len(A)
    def sort(left, right):
        if right - left <= 1:
            return
        mid = (left + right) // 2
        sort(left, mid)
        sort(mid, right)
        i = left
        j = mid
        k = left
        while i < mid and j < right:
            if A[i] <= A[j]:
                temp[k] = A[i]
                i += 1
            else:
                temp[k] = A[j]
                j += 1
            k += 1
        while i < mid:
            temp[k] = A[i]
            i += 1
            k += 1
        while j < right:
            temp[k] = A[j]
            j += 1
            k += 1
        for t in range(left, right):
            A[t] = temp[t]
    sort(0, len(A))
    return A

A = [5, 3, 8, 1, 2, 7]
merge_sort(A)
print(A)

# your algorithm time complexity is:Θ(nlogn)
````

idea:最直观的方法是尝试所有可能的买入和卖出日期的组合。可以使用两层嵌套循环：外层循环遍历所有的买入天数i，内层循环遍历所有合法的卖出天数j（j>=i）。 在内层循环中，我们累加从第i天到第=j天的收益，并记录下曾经出现过的最大收益以及对应的买卖日期。
时间复杂度计算：
外层循环执行$n$次，内层循环平均执行$n/2$次。 因此总的计算次数约为$n^2 / 2$，时间复杂度为$O(n^2)$。 

````python
def max_profit_basic(profits):
    n = len(profits)
    max_sum = float('-inf')
    buy_day = 0
    sell_day = 0
    for i in range(n):
        current_sum = 0
        for j in range(i, n):
            current_sum += profits[j]
          
            if current_sum > max_sum:
                max_sum = current_sum
                buy_day = i + 1   
                sell_day = j + 1      
    return max_sum, buy_day, sell_day
def max_profit_optimized(profits):
    if not profits:
        return 0, 0, 0
    max_sum = float('-inf')  
    current_sum = 0          
    buy_day = 0              
    sell_day = 0             
    temp_buy = 0             
    for i in range(len(profits)):
        if current_sum <= 0:
            current_sum = profits[i]
            temp_buy = i
        else:
            current_sum += profits[i]
        if current_sum > max_sum:
            max_sum = current_sum
            buy_day = temp_buy + 1  
            sell_day = i + 1
    return max_sum, buy_day, sell_day
if __name__ == "__main__":
    profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
    max_profit, buy, sell = max_profit_optimized(profits)
    
    print(f"profits={profits}")
    print(f"第{buy}天买入，第{sell}天卖出，收益最大：{max_profit}")
# your algorithm time complexity is: O(n)
````

1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:![问题1的推导过程](./fig/3.jpg)·

````python
def find_one_local_min(arr):
    
    n = len(arr)
    low, high = 0, n - 1  

    while low <= high:
        mid = (low + high) // 2
       
        left_ok = (mid == 0) or (arr[mid] < arr[mid - 1])
        right_ok = (mid == n - 1) or (arr[mid] < arr[mid + 1])
        if left_ok and right_ok:
            return arr[mid]
        if mid > 0 and arr[mid - 1] < arr[mid]:
            high = mid - 1
        else:
            low = mid + 1
    return None

def find_all_local_min(arr):
   
    n = len(arr)
    res = []
    for i in range(1, n - 1):  
        if arr[i] < arr[i - 1] and arr[i] < arr[i + 1]:
            res.append(arr[i])
    return res

if __name__ == "__main__":
    A = [9, 3, 7, 2, 1, 4, 5]  
    print("数组:", A)
    print("一个局部最小:", find_one_local_min(A))
    print("所有局部最小:", find_all_local_min(A))
    # your algorithm time complexity is: O(n)
````

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:1.排序后取前k个,先对整个数组进行排序，然后取排序后数组的前k个元素。由于元素互不相同，直接排序后输出即可。步骤：使用高效的排序算法对数组进行升序排序，返回排序后数组的前k个元素。时间复杂度：排序：O(nlogn)）。取前k个：O(k)。总：O(nlogn + k) = O(nlogn)。
2.建堆并k次提取最小值，利用最小堆（优先队列）的性质。先将所有n个元素插入最小堆，然后连续调用k次extract_min，弹出的序列即为前k小的数，且按升序输出。总时间复杂度：O(nlogn+klogn)=O((n+k)logn)。如果k接近n，则为O(nlogn)；如果k << n，仍为O(nlogn)。
3.当数据总量超过内存，无法一次性全部读入时，我们可以维护一个大小为k的最大堆。遍历数据流时：若堆的大小小于k，直接插入；若当前元素比堆顶小，则替换堆顶。遍历完成后，堆中保留的就是所有数据中最小的k个数。最后对堆内元素排序即可得到升序结果。
处理每个元素：若堆未满，插入O(logk)；若需要替换，heapreplace也是O。
总共n个元素：O(nlogk)。
最后对k个元素排序：O(klogk)。
总：O(nlogk+klogk)=O(nlogk)。



````python
import heapq
def find_k_smallest_by_sorting(arr, k):
    if k <= 0:
        return []
    sorted_arr = sorted(arr)        
    return sorted_arr[:k]

def find_k_smallest_by_heap_insert(arr, k):
    if k <= 0:
        return []
    heap = []
    for val in arr:
        heapq.heappush(heap, val)       
    result = []
    for _ in range(min(k, len(heap))):
        result.append(heapq.heappop(heap))  
    return result

def find_k_smallest_stream(data_stream, k):
    if k <= 0:
        return []
    max_heap = []
    for val in data_stream:
        if len(max_heap) < k:
            heapq.heappush(max_heap, -val)          
        else:
            if val < -max_heap[0]:                  
                heapq.heapreplace(max_heap, -val) 
    result = sorted([-x for x in max_heap])
    return result
# your algorithm time complexity is: O(n log k)

if __name__ == "__main__":
    arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
    k = 3
    print("方法1（排序）：", find_k_smallest_by_sorting(arr, k))
    print("方法2（堆插入）：", find_k_smallest_by_heap_insert(arr, k))
    print("方法3（流式处理）：", find_k_smallest_stream(iter(arr), k))

  
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:星球在同一直线上，速度相同，正数向右运动，负数向左运动。当两个星球方向相对时才会相撞。因为速度相同，碰撞只会发生在相邻的星球之间，且碰撞后可能导致新的相邻对。
可以使用栈来模拟这个过程，从左到右遍历数组：
遇到正数时，直接压入栈中。
遇到负数时，需要检查栈顶：
如果栈为空或栈顶为负数，说明当前星球左边没有向右运动的星球，直接压入栈。
如果栈顶为正数，则可能发生碰撞。比较绝对值：
若当前星球质量更大，则栈顶星球消失，继续与新的栈顶比较。
若当前星球质量更小，则当前星球消失，停止比较。
若质量相等，则当前星球消失，栈顶星球保留，停止比较。
遍历结束后，栈中剩余元素即为最终存活的星球，顺序与原始顺序一致。
每个星球最多入栈一次、出栈一次，因此总操作次数为 O(n)。所以时间复杂度为 O(n)。


````python
def asteroid_collision(asteroids):
    stack = []
    for a in asteroids:
        if a > 0:
            stack.append(a)
        else:
            while stack and stack[-1] > 0:
                # 比较绝对值
                if stack[-1] > -a:      
                    break              
                elif stack[-1] < -a:    
                    stack.pop()        
                else:                    
                    break
            else:
                stack.append(a)
    return stack

if __name__ == "__main__":
    test_cases = [
        ([-3, -6, 2, 8, 5, -8, 9, -2, 1], [-3, -6, 2, 8, 9, 1]),
        ([23, -8, 9, -3, -7, 9, -23, 22], [23, 22]),
    ]
    for arr, expected in test_cases:
        result = asteroid_collision(arr)
        print(f"输入: {arr}\n输出: {result}")
       
        # your algorithm time complexity is: O(n)
````

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:![问题1的推导过程](./fig/4.jpg)·

````python
def knapsack_01(items, capacity):
    dp = [0] * (capacity + 1)
    n = len(items)
    choice = [[False] * (capacity + 1) for _ in range(n)]
    for i in range(n):
        weight, value = items[i]
        
        for j in range(capacity, weight - 1, -1):
            if dp[j - weight] + value > dp[j]:
                dp[j] = dp[j - weight] + value
                choice[i][j] = True
    selected_items = []
    j = capacity
    for i in range(n - 1, -1, -1):
        if choice[i][j]:
            selected_items.append(items[i])
            j -= items[i][0] 
        
    return dp[capacity], selected_items[::-1]

marbles = [(5, 10), (4, 40), (6, 30), (3, 50)]
cart_capacity = 9

max_val, selected = knapsack_01(marbles, cart_capacity)
print(f"最大承重: {cart_capacity}")
print(f"小推车能装入的最大价值为: {max_val}")
print(f"选择的大理石 (重量, 价值): {selected}")
#your algorithm time complexity is: O(n*W)
````

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:二叉搜索树的中序遍历序列一定是一个严格递增的序列。 如果有两个节点被交换了，中序遍历时就会出现前一个节点的值大于当前节点值的逆序情况。
我们可以进行一次中序遍历，维护一个prev指针记录上一个节点，当前节点为current。 如果发现prev的值大于current的值：
如果是第一次遇到逆序对，说明prev就是第一个错误节点，我们把 prev 记录下来，同时把current也先记录为第二个错误节点的候选。
如果是第二次遇到逆序对，说明current才是真正的第二个错误节点，此时我们更新第二个错误节点的记录。 遍历结束后，交换这两个记录下来的节点的值即可。

````python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def recoverTree(self, root: TreeNode) -> None:
        self.first = None
        self.second = None
        self.prev = None
        def inorder(node):
            if not node:
                return
            inorder(node.left)
            if self.prev and self.prev.val > node.val:
                if not self.first:
                    self.first = self.prev
                    self.second = node
                else:
                    self.second = node
            self.prev = node
            inorder(node.right)
        inorder(root)
        if self.first and self.second:
            self.first.val, self.second.val = self.second.val, self.first.val

def tree_to_array(root):
    if not root:
        return []
    result = []
    queue = [root]
    while queue:
        node = queue.pop(0)
        if node:
            result.append(node.val)
            queue.append(node.left)
            queue.append(node.right)
        else:
            result.append(None)
    while result and result[-1] is None:
        result.pop()
    return [val if val is not None else "null" for val in result]

if __name__ == "__main__":
    sol = Solution()
    root1 = TreeNode(1)
    root1.left = TreeNode(3)
    root1.left.right = TreeNode(2)
    print("输入: root =", tree_to_array(root1))
    sol.recoverTree(root1)
    print("输出:", tree_to_array(root1))
    root2 = TreeNode(3)
    root2.left = TreeNode(1)
    root2.right = TreeNode(4)
    root2.right.left = TreeNode(2)
    print("输入: root =", tree_to_array(root2))
    sol.recoverTree(root2)
    print("输出:", tree_to_array(root2))
#your algorithm time complexity is: O(n)
````

具体步骤：

右指针right，初始指向数组的最右端。
左侧最大值left_max，用于记录从左向右扫描到的最高柱子。
右侧最大值right_max，用于记录从右向左扫描到的最高柱子。
在左指针和右指针相遇之前，我们循环执行以下判断：
首先更新left_max和right_max，确保它们始终是当前指针及其外侧的最高高度。
接着，比较left_max 和right_max 的大小：
如果left_max小于right_max：这意味着对于左指针所在的位置，它的短板已经确定在左侧了。 此时不用管右侧到底有多高，左指针位置能接的水量就是left_max减去它本身的高度。 将这部分水量累加到总和中，并将左指针向右移动一步。
如果left_max大于或等于right_max：这意味着对于右指针所在的位置，它的短板确定在右侧。 同理，右指针位置能接的水量就是 right_max 减去它本身的高度。 累加水量后，将右指针向左移动一步。
当左右指针相遇时，整个数组扫描完毕，返回累加的总水量即可。时间复杂度：O（n）其中n是柱子高度数组的长度。 左指针和右指针最多把整个数组遍历一次，没有嵌套循环，因此时间复杂度是线性的。

````python
def trap(height):
   
    if not height or len(height) < 3:
        return 0
    left = 0
    right = len(height) - 1
    left_max = height[left]
    right_max = height[right]
    water_trapped = 0
    while left < right:
        if left_max < right_max:
           
            left += 1
            left_max = max(left_max, height[left])
          
            water_trapped += left_max - height[left]
        else:
            right -= 1
            right_max = max(right_max, height[right])
            water_trapped += right_max - height[right]
    return water_trapped

if __name__ == "__main__":
    height1 = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
    print(f"输入: {height1}")
    print(f"输出: {trap(height1)} ")
    height2 = [4, 2, 0, 3, 2, 5]
    print(f"输入: {height2}")
    print(f"输出: {trap(height2)} ")
    #your algorithm time complexity is: O(n)
````

## 问题 10

给你一个二叉树的根节点 root ，树中每个节点都存放有一个 0 到 9 之间的数字。每条从根节点到叶节点的路径都代表一个数字：

> 例如，从根节点到叶节点的路径 1 -> 2 -> 3 表示数字 123 。

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

e.g.:
---
1.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-1.jpg"> </div>

 输入：root = [1,2,3], 输出：25

解释：从根到叶子节点路径 1->2 代表数字 12，1->3 代表数字 13，因此和为 12 + 13 = 25 。

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

idea：在遍历二叉树的过程中，可以自顶向下传递一个当前路径上已经累积的数字。每当访问到一个新的节点时，就把之前传递下来的累积数字乘以10，再加上当前节点自身的值，从而形成最新的累积数字。
接着判断当前节点的状态：如果当前节点是叶子节点（也就是既没有左孩子也没有右孩子），说明我们已经完整地走完了一条路径，此时直接返回这个刚刚算出的累积数字即可。
如果当前节点不是叶子节点，我们就继续向下递归，分别计算它的左子树和右子树中所有路径的数字之和，然后将这两部分的结果相加，作为当前节点向上一层返回的结果。
时间复杂度：O（N） 其中N是这棵二叉树的节点总数。在深度优先搜索的过程中，整棵树里的每一个节点都会被精确地访问一次，所有的计算也都是常数时间的，因此总的时间复杂度是线性的。

````python

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

class Solution:
    def sumNumbers(self, root: TreeNode) -> int:
        def dfs(node, current_sum):
            if not node:
                return 0
            current_sum = current_sum * 10 + node.val
            if not node.left and not node.right:
                return current_sum
            left_sum = dfs(node.left, current_sum)
            right_sum = dfs(node.right, current_sum)
            return left_sum + right_sum
        return dfs(root, 0)
    
if __name__ == "__main__":
    sol = Solution()
    root1 = TreeNode(1)
    root1.left = TreeNode(2)
    root1.right = TreeNode(3)
    result1 = sol.sumNumbers(root1)
    print("输出:", result1)
    root2 = TreeNode(4)
    root2.left = TreeNode(9)
    root2.right = TreeNode(0)
    root2.left.left = TreeNode(5)
    root2.left.right = TreeNode(1)
    result2 = sol.sumNumbers(root2)
    print("输出:", result2)
      #your algorithm time complexity is: O(n)
````
